# 데이터 전처리
**1. 불필요한 데이터 제거 : id 제거**

**2. 데이터 타입 변경 : target 인코딩 (문자 -> 숫자)**

**3. 학습 데이터, 평가 데이터 분리**

```python
# id 제거
train = train.drop(['id'], axis = 1)
test = test.drop(['id'], axis = 1)
```

```python
# target encoding

# 람다식을 사용해 dictionary 자료형으로 대치
# dic = {'Class_1':1,'Class_2':2, 'Class_3':3, 'Class_4':4, 'Class_5':5,
#       'Class_6':6, 'Class_7':7, 'Class_8':8, 'Class_9':9}
# encoding_target = train['target'].apply(lambda x: dic[x])

# label encoding 사용
le = LabelEncoder()
le = le.fit(train['target'])
target = le.transform(train['target'])
```

```python
# 학습 데이터, 평가 데이터 분리

# 차집합 사용
# x_col = list(train.columns.difference(['target']))
# x = train[x_col]

x = train[train.columns[0:-1]]
y = target

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 39)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# 학습 및 평가

```python
rf_clf = RandomForestClassifier(random_state = 39)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)

print("랜덤 포레스트 정확도 : {0:.4f}".format(accuracy_score(y_test, pred)))
```

```python
# kaggle에서 원하는 정확도 지표 - logloss(다분류)

def logloss(answer_array, proba_array):

    # 0이면 0에 가까운 값으로 치환
    MIN_VALUE = 1e-15

    # array의 크기를 가져옵니다.
    size = answer_array.shape[0]

    # 반복문으로 logloss의 합 계산
    logloss_sum = 0

    for answer, arr in zip(answer_array, proba_array):
        # 예측값에 대한 확률 array의 index는 (label encoding한 target값) - 1
        proba = arr[answer - 1]

        # 0이면  0에 가까운 값으로 치환
        if proba <= MIN_VALUE:
            proba = MIN_VALUE

        # 음의 로그함수에 넣어서 logloss 계산
        logloss_sum += -np.log(proba)

    # logloss의 평균 계산
    result = logloss_sum / size

    return result

proba = rf_clf.predict_proba(X_test) # 예측 확률 불러오기
logloss_result = logloss(y_test, proba)
print('logloss : {0}'.format(logloss_result))
```

# 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators' : [30, 60, 90, 120, 150, 180],
    'max_depth' : [10, 20, 30, 40, 50, 60]
}

rf_clf_tunning = RandomForestClassifier(random_state=39, n_jobs=-1)
# n_jobs : 학습 수행을 위해 CPU 코어를 병렬적으로 활용할 작업 수, -1은 모두 사용
#          코어 개수보다 크게 지정하는 것은 도움되지 않음
grid_rf1 = GridSearchCV(rf_clf_tunning, param_grid = params, cv = 2, n_jobs = -1)
# refit default가 true라 GridSearchCV 작동 후에 자동으로 해당 파라미터로 학습된 모델 return
grid_rf1.fit(X_train, y_train)

print("최적 파라미터 : ",grid_rf1.best_params_)
print("최고 예측 정확도 : {0:.5f}".format(grid_rf1.best_score_))
```

```python
params = {
    'n_estimators' : [50, 100, 150, 200],
    'max_depth' : [40, 42, 44, 46, 48, 50]
}

rf_clf_tunning = RandomForestClassifier(random_state=39, n_jobs=-1)
grid_rf2 = GridSearchCV(rf_clf_tunning, param_grid = params, cv = 2, n_jobs = -1)
grid_rf2.fit(X_train, y_train)

print("최적 파라미터 : ",grid_rf2.best_params_)
print("최고 예측 정확도 : {0:.5f}".format(grid_rf2.best_score_))
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# 다른 평가지표 계산
def clf_eval(y_test, pred):
    # confusion = confusion_matrix(y_test, pred)  # 오차행렬
    accuracy = accuracy_score(y_test, pred)     # 정확도
    precision = precision_score(y_test, pred, average='macro')   # 정밀도
    recall = recall_score(y_test, pred, average='macro')         # 재현율
    f1 = f1_score(y_test, pred, average='macro')

    print('''F1 스코어 : {0:.4f}
정확도 : {1:.4f}
정밀도 : {2:.4f}
재현율 : {3:.4f}'''.format(f1, accuracy, precision, recall))


clf_eval(y_test, pred2)
```
